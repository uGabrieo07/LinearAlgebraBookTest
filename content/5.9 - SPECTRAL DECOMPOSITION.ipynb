### Spectral Decomposition

[The previous notebook](https://martinamj.github.io/Online-Linear-Algebra/lab/index.html?path=5.8+-+SYMMETRIC+MATRICES+AND+SCHUR+THEOREM.ipynb) 
indeed states that Schur’s decomposition of a symmetric matrix takes the
form
$$\mathbf{A}=\mathbf{P}^{\rm T}\boldsymbol{\Lambda}\mathbf{P}=\sum_{i=1}^n\lambda_i\mathbf{v}_i\mathbf{v}_i^{\rm T}$$
where $\boldsymbol{\Lambda}$ is diagonal and contains the eigenvalues of
$\mathbf{A}$, and $\mathbf{P}^{\rm T}$ is orthogonal with the
orthonormal eigenvectors on its columns. We can now relate Schur’s
decomposition applied to symmetric matrices, or self-adjoint operators,
as defined in the first chapter. We will state the relationship as a
theorem.

(Spectral Decomposition)<span id="theo:spectral_decomposition"
label="theo:spectral_decomposition"></span> Let
$\mathbf{A}:\mathbb{R}^n\rightarrow\mathbb{R}^n$ be a linear operator in
$\mathcal{U}$ induced by the square matrix $\mathbf{A}$. The linear
operator $\mathbf{A}$ is self-adjoint if, and only if, matrix
$\mathbf{A}=\mathbf{P}^{\rm T}\boldsymbol{\Lambda}\mathbf{P}$, where
$\mathbf{P}$ is orthogonal and $\boldsymbol{\Lambda}$ is diagonal.

*Proof.* From [Schur's Theorem](https://martinamj.github.io/Online-Linear-Algebra/lab/index.html?path=5.8+-+SYMMETRIC+MATRICES+AND+SCHUR+THEOREM.ipynb),
there exist $\mathbf{P}$ orthogonal and $\mathbf{T}$ triangular such
that $\mathbf{P}\mathbf{A}\mathbf{P}^{\rm T}=\mathbf{T}$. Therefore,
$\mathbf{A}=\mathbf{P}^{\rm T}\mathbf{T}\mathbf{P}$. Similarly,
$\mathbf{A}^{\rm T}=\mathbf{P}^{\rm T}\mathbf{T}^{\rm T}\mathbf{P}$. If
$\mathbf{A}$ is self-adjoint, then
$\mathbf{A}=\mathbf{A}^{\rm T}\Rightarrow\mathbf{T}=\mathbf{T}^{\rm T}$
because $\mathbf{P}$ is orthogonal. But as $\mathbf{T}$ is triangular,
then it must be diagonal. On the other hand, if
$\mathbf{T}=\boldsymbol{\Lambda}$, diagonal, then
$\mathbf{T}=\mathbf{T}^{\rm T}\Rightarrow\mathbf{A}=\mathbf{A}^{\rm T}$,
i.e., $\mathbf{A}$ is self-adjoint.

The next theorem presents an important property of the eigenvalues of
symmetric matrices.

<span id="theo:real_eigenvalues_symm_matrices"
label="theo:real_eigenvalues_symm_matrices"></span> Let
$\mathbf{A}:\mathbb{R}^n\rightarrow\mathbb{R}^n$ be a linear operator
induced by a symmetric matrix $\mathbf{A}$. The eigenvalues of
$\mathbf{A}$ are real.

*Proof.* Let $\lambda$ and $\mathbf{v}$ be an eigenvalue and eigenvector
pair of $\mathbf{A}$: $\mathbf{A}\mathbf{v}=\lambda\mathbf{v}$. The
equality is maintained for the complex conjugate, therefore
$\mathbf{A}\mathbf{v}^*=\lambda^*\mathbf{v}^*$. If we multiply from the
left the first equation by $(\mathbf{v}^*)^{\rm T}$ and the second
equation by $\mathbf{v}^{\rm T}$, we get
$(\mathbf{v}^*)^{\rm T}\mathbf{A}\mathbf{v}=\lambda(\mathbf{v}^*)^{\rm T}\mathbf{v}=\lambda\|\mathbf{v}\|_2^2$
and
$\mathbf{v}^{\rm T}\mathbf{A}\mathbf{v}^*=\lambda^*\mathbf{v}^{\rm T}\mathbf{v}^*=\lambda^*\|\mathbf{v}\|_2^2$,
respectively. As these terms are scalars, they are equal to their
transposed quantities. Therefore the second term can be rewritten as
$(\mathbf{v}^*)^{\rm T}\mathbf{A}^{\rm T}\mathbf{v}=\lambda^*\|\mathbf{v}\|_2^2$.
As $\mathbf{A}$ is symmetric, i.e., $\mathbf{A}=\mathbf{A}^{\rm T}$, and
$\|\mathbf{v}\|_2^2\ne0$ the two terms are equal and
$\lambda=\lambda^*$.

### Example

Although the spectral decomposition can be easily proved and seems to be just another matrix decomposition algorithm, it proves itself to be a powerful tool in data analysis and machine learning. One of the main applications of spectral decomposition in Dimensionality Reduction and Principal Component Analysis (PCA). 

In data analysis, it is common to use databases that contain many features. "Feature" is data science jargon that refers to a variable (or parameter) of the space in which the data belongs. A database that contains many feature belongs to a space that has many variables, i.e has many degrees of freedom. That means it has a high dimensionality. 

While a high dimensionality may lead to a better described problem, it can also be a challenge in some aspects. Sometimes, the data varies in a direction that is a linear combination of the features. This makes it dificult to analise which features are most significant to the database. Other times, features are closely related, but are not linearly dependent to eachother. This means that those features are *highly collinear*. High collinearity often causes problems in finding prediction models, making it difficult to find a unique solution. 

In such context, an adequate solution is to find the features with the highest variability and independence. In other words, with maximized covariance and orthogonal directions. 

Covariance is a value that quantifies the relation between two random variables. It is given by the expression: $cov(X,Y) = E[(X-E[X])(Y-E[Y])^T]$, in which $X$ and $Y$ are vectors. In this case, the random variables (vectors) are the each features values for all samples. It is possible to define a covariance matrix, in which each row and column refer to a feature. In order for this matrix to be symmetric, the row and column of same index refer to the same feature. With a symmetric and real matrix, we may apply spectral decomposition. 

The diagonal matrix of eigenvalues is the focus of our analysis. If we multiply a decomposed matrix, we obtain the following equation: $\begin{bmatrix} a & b \\ c & d \end{bmatrix} \cdot \begin{bmatrix} e_1 & 0 \\ 0 & e_2 \end{bmatrix} \cdot \begin{bmatrix} a & c \\ b & d \end{bmatrix} = \begin{bmatrix} a^2e_1+b^2e_2 & ace_1+bde_2 \\ ace_1+bde_2 & c^2e_1+d^2e_2 \end{bmatrix}$

Notice that the eigenvalues $e_1$ and $e_2$ only multiply values in the first column and in the second column, respectively. If one of those eigenvalues happens to be close to zero, that means that the column that it is associated with is not relevant in the construction of the covariance matrix. Therefore, removing the line and columns of same index as that eigenvalue will not greatly affect the remaining covariances in the matrix. 

In conclusion, we find that the K highest eigenvalues will give us the K principal components. In agreement with that, the eigenvectors that are associated with the highest eigenvalues will be in the orthogonal directions in which values experience the highest dispersion. We then use these eigenvectors as a new basis for the space in which the data is located. 

In the example below, we will observe how applying PCA can improve the quality of a predictive model's performance. 

We will first illustrate how PCA changes the space of a dataset. We will use the Iris data set (fist used in [3.1](https://martinamj.github.io/Online-Linear-Algebra/lab/index.html?path=3.1+-+METRICS%2C+DISTANCES+AND+NORMS.ipynb)). This data set originally has 4 features (sepal length, sepal width, petal length, petal width), which is why we need two 2x2 plots to visualize the data. PCA will be applied in the data set in order to entirely describe the data set suing only two principal components.

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import pandas as pd

#Loading iris dataset
iris = load_iris()
X = iris.data
y = iris.target
target_names = iris.target_names

#Loading PCA and creating new data
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

#Creating a dataframe, adding target column
df = pd.DataFrame(X_pca, columns=["PC1", "PC2"])
df["species"] = [target_names[i] for i in y]

fig = plt.figure(layout='tight', figsize=(10,4))
grid_spec = fig.add_gridspec(4,4)
f_ax1 = fig.add_subplot(grid_spec[:-2,:2])
f_ax1.scatter(iris.data[:, 0], iris.data[:, 1], s=7, c=iris.target)
f_ax1.set_xlabel(iris.feature_names[0])
f_ax1.set_ylabel(iris.feature_names[1])
f_ax2 = fig.add_subplot(grid_spec[-2:,:2])
f_ax2.scatter(iris.data[:, 2], iris.data[:, 3], s=7, c=iris.target)
f_ax2.set_xlabel(iris.feature_names[2])
f_ax2.set_ylabel(iris.feature_names[3])
f_ax3 = fig.add_subplot(grid_spec[0:,2:])
scatter_pca = f_ax3.scatter(df["PC1"], df["PC2"], s=7, c=iris.target)
f_ax3.set_xlabel("Principal Component 1")
f_ax3.set_ylabel("Principal Component 2")
f_ax3.set_title("PCA on Iris Data Set")

fig = f_ax3.legend(scatter_pca.legend_elements()[0], iris.target_names, loc="lower right")

plt.show()
```

As you can see, after PCA, it is possible to visualize the data entirely in 2 dimensions, which are orthogonal to eachother. It is also clear to see that the data is more disperse. This makes it easier for prediction model algorithms to find a stable model. We will test this using a K-nearest neighbors model. 

The quality of a solution can be quantified by error metrics. In this example, we will used mean square error (MSE) and $R^2$ score. These metrics are, respectively: $\text{MSE} = \frac{1}{N} \sum_{i=1}^{N} (y_i - \hat{y}_i)^2$ and $R^2 = 1 - \frac{\sum_{i=1}^{N} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{N} (y_i - \bar{y})^2}$

In the formulas above, $\hat{y}_i$ refers to the expected value of index $i$ of a database and $\bar{y}$ refers to the mean of the same database. Therefore, MSE accounts the mean of the square of differences (residual) between each expected value and predicted value. We can conclude that the closer this number is to 0, the better our predictions are. Similarly, in $R^2$ score, the closer the fraction is to 0 the better our predictions are, so, the $R^2$ score of a particular prediction must be closest to 1. 

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

X_train_iris, X_test_iris, y_train_iris, y_test_iris = train_test_split(X, y, test_size=0.33, random_state=42)
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(X_pca, y, test_size=0.33, random_state=42)

knn = KNeighborsClassifier(n_neighbors=max(1, 6))
knn.fit(X_train_iris, y_train_iris)
result_iris = knn.predict(X_test_iris)
r2_iris = r2_score(y_test_iris, result_iris)
mse_iris = mean_squared_error(y_test_iris, result_iris)

knn_pca = KNeighborsClassifier(n_neighbors=max(1, 6))
knn_pca.fit(X_train_pca, y_train_pca)
result_pca = knn_pca.predict(X_test_pca)
r2_pca = r2_score(y_test_pca, result_pca)
mse_pca = mean_squared_error(y_test_pca, result_pca)

print(f"R2 score: \nOriginal: {r2_iris} and PCA: {r2_pca}")
print(f"MSE: \nOriginal: {mse_iris} and PCA: {mse_pca}")
```

As it is possible to see, applying the PCA dataset to the KNN model produces better results. In this example, however, the differences between the error metrics are almost negligible. This is due to the fact that the dataset we used to illustrate this example originally has a low dimensionality. PCA is best applied in datasets with many features. Yet, even in this simple example, it is possible to visualize how it can be used positively.
